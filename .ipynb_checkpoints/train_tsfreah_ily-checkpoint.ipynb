```python
import pandas as pd
import matplotlib.pyplot as plt
from tsfresh import extract_features
plt.style.use('seaborn')

import numpy as np
import datetime
plt.rcParams['figure.dpi'] = 120
plt.rcParams['figure.figsize'] = (12, 5)
```

# Загружаем DataFrame

```python
df = pd.read_csv('daily-min-temperatures.csv')
df
```

```python
df.columns = ['date', 'temp']
```

```python
df
```

```python
df_extract = extract_features(df, column_id='date', n_jobs=4)
```

```python
df_extract
```

#### Избавляюсь от столбцов с NaN, пока что просто удаляю. В дальнейшем добавлю заполнение значений. А также удалила столбцы, где одни нули

```python
df_extract = df_extract.dropna(axis=1, how='any')
```

```python
df_extract = df_extract.loc[:, (df_extract != 0).any(axis=0)]
```

```python
df_extract
```

### Использую разбиение TimeSeriesSplit

```python
from sklearn.model_selection import TimeSeriesSplit
```

```python
tss = TimeSeriesSplit(6)
```

```python
y = df.temp
```

```python
X = df_extract.values
```

## Обучаю дерево, заполняю массивы с MSE, чтобы получить график ошибок

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
rf = RandomForestRegressor(n_estimators=200)
```

```python
rf_train_mse = []
rf_test_mse = []

for train_index, test_index in tss.split(X, y):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    rf.fit(X_train, y_train)
    
    rf_train_mse.append(mean_squared_error(y_train, rf.predict(X_train)))
    rf_test_mse.append(mean_squared_error(y_test, rf.predict(X_test)))
    
```
